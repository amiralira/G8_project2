```python
import numpy as np
import pandas as pd
```

```python
mdf = pd.read_excel('lap_data.xlsx')
mdf
```

```python
gpu_col = mdf['GPU']
gpu_flag = pd.Series(np.zeros((len(gpu_col))))

ct = 0
for i in gpu_col:
    if 'Intel' not in i:
        gpu_flag[ct] = 1

    ct += 1
gpu_flag.value_counts()
```

```python
gpu_flag = pd.DataFrame(gpu_flag)
gpu_flag['pd_id'] = mdf['product_id']
gpu_flag
```

```python
gpu_flag.to_csv('G_FLAG.csv', index=False)
```
