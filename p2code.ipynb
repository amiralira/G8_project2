```python
first_file
first_file['Order_Date'] = first_file['Order_Date'].str.split('-').str[0]
first_file
```

```python
first_file = first_file.rename(columns={'Order_Date': 'year'})
first_file
```

```python
usd_tooman
```

```python
usd_tooman_by_year= usd_tooman.groupby('date_pr',as_index=False)['price_avg'].mean()
usd_tooman_by_year=usd_tooman_by_year.rename(columns={'date_pr':'year'})
usd_tooman_by_year['price_avg']=usd_tooman_by_year['price_avg']/10
usd_tooman_by_year
```

```python

merged_df = first_file.merge(usd_tooman_by_year, on='year', how='inner')
merged_df['price(dollar)'] = merged_df['Price'] / merged_df['price_avg']
merged_df
```

```python

correlation = merged_df['price(dollar)'].corr(merged_df['Quantity'])

# Display the correlation coefficient
print(f"Correlation coefficient: {correlation}")
```

```python
merged_df=merged_df.rename(columns={'price(dollar)':'priced'})

merged_df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('priced ~ Category', data=merged_df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
column_to_keep=['Weight','priced']
```

```python
import numpy as np
new_df = merged_df[column_to_keep]
new_df
new_df['Weight'] = new_df['Weight'].apply(lambda x: float(x.rstrip('kg').rstrip('kgs')))

new_df
correlation = np.corrcoef(new_df['Weight'], new_df['priced'])[0, 1]

# Check the strength and direction of the correlation
if correlation > 0:
    print("There is a positive correlation between weight and price.")
elif correlation < 0:
    print("There is a negative correlation between weight and price.")
else:
    print("There is no linear correlation between weight and price.")
print(correlation)    
```

```python
def extract_cpu_brand(cpu_description):
    if 'Intel' in cpu_description:
        return 'Intel'
    elif 'AMD' in cpu_description:
        return 'AMD'
    else:
        return 'Other'

# Apply the extract_brand function to create a new 'brand' column
merged_df['CPU_brand'] = merged_df['CPU'].apply(extract_cpu_brand)

# Group the data by brand and calculate the average price for each brand
brand_prices = merged_df.groupby('CPU_brand')['priced'].mean()

# Display the average prices for each brand
print(brand_prices)
```

```python
cpu_demand = merged_df.groupby('CPU_brand')['Quantity'].count()
cpu_demand
```

```python
demand_data =merged_df.groupby(['year', 'CPU_brand'])['Quantity'].sum().unstack(fill_value=0)
import matplotlib.pyplot as plt

# Plot the demand for each brand over time
demand_data.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Demand')
plt.title('Demand for CPU Brands Over Time')
plt.legend(title='Brand')
plt.show()
```
